# Task â„–1

```python
import numpy as np
import matplotlib.pyplot as plt
import mplcyberpunk

plt.style.use("cyberpunk")

def phi(t, mu, sigma):
  return np.exp(-sigma**2 * t**2 / 2)

# ÐŸÐ°Ñ€Ð°Ð¼ÐµÑ‚Ñ€Ð¸
mu = 0
sigma = 1

# Ð“Ñ€Ð°Ñ„Ñ–ÐºÐ¸
t = np.linspace(-10, 10, 1000)
phi_mu0_sigma1 = phi(t, mu, sigma)
phi_mu1_sigma1 = phi(t, 1, sigma)
phi_mu0_sigma2 = phi(t, mu, 2*sigma)

plt.plot(t, phi_mu0_sigma1, label="$\mu = 0, \sigma = 1$")
plt.plot(t, phi_mu1_sigma1, label="$\mu = 1, \sigma = 1$")
plt.plot(t, phi_mu0_sigma2, label="$\mu = 0, \sigma = 2$")

plt.xlabel("$t$")
plt.ylabel("$\phi(t)$")
plt.legend()
mplcyberpunk.add_glow_effects()
plt.show()
```

# Task â„–3

```python
def gauss_density(x, mu, sigma):
    return 1 / (np.sqrt(2 * np.pi * sigma**2)) * np.exp(-(x - mu)**2 / (2 * sigma**2))

x = np.linspace(-10, 10, 1000)

plt.plot(x, gauss_density(x, 0, 1))
plt.xlabel("x")
plt.ylabel("p(x)")
plt.legend("Gauss density")
mplcyberpunk.add_glow_effects()
plt.show()

def gauss_characteristic_function(t):
    return np.exp(-t**2 / 2)

t = np.linspace(-10, 10, 1000)

plt.plot(t, gauss_characteristic_function(t))
plt.xlabel("t")
plt.ylabel("Ï•(t)")
plt.legend("Gauss characteristic function")
mplcyberpunk.add_glow_effects()
plt.show()
```

# Task â„–4

```python
import math

def poisson(a, b, n):
  return (a * b)**n * np.exp(-(a + b)) / math.factorial(n)

x = np.arange(0, 100)
y = [poisson(2, 3, n) for n in x]

plt.plot(x, y)
plt.xlabel("n")
plt.ylabel("p(n)")
mplcyberpunk.add_glow_effects()
plt.show()
```

# Task â„–5

```python
def f(x):
    return 1 / (2 * np.pi) * np.arctan(x / np.sqrt(2))

x = np.array([-2, -1, 0, 1, 2])

f_x = f(x)

print(f_x)

plt.plot(x, f_x)
plt.xlabel("x")
plt.ylabel("f(x)")
mplcyberpunk.add_glow_effects()
plt.show()
```

# Task â„–6

```python
def f(x, lamb):
    return lamb / 2 * np.exp(-lamb * np.abs(x))

x = np.linspace(-5, 5, 1000)

plt.plot(x, f(x, 0.5), label="Î» = 0.5")
plt.plot(x, f(x, 1), label="Î» = 1")
plt.plot(x, f(x, 2), label="Î» = 2")

plt.legend()
mplcyberpunk.add_glow_effects()
plt.show()
```

# Task â„–7

```python
def p(x, lamb, sigma):
    return 1 / (np.sqrt(2 * np.pi * sigma**2) * x) * np.exp(-(np.log(x) - lamb)**2 / (2 * sigma**2))

x = np.linspace(0, 10, 100)

plt.plot(x, p(x, 1, 1), label="lambda = 1, sigma = 1")
plt.plot(x, p(x, 2, 1), label="lambda = 2, sigma = 1")
plt.plot(x, p(x, 1, 2), label="lambda = 1, sigma = 2")

plt.xlabel("x")
plt.ylabel("p(x)")
plt.legend()
mplcyberpunk.add_glow_effects()
plt.show()
```
