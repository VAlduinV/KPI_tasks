# Task â„–1

```python
import numpy as np
import matplotlib.pyplot as plt
import mplcyberpunk

plt.style.use("cyberpunk")

def phi(t, mu, sigma):
  return np.exp(-sigma**2 * t**2 / 2)

# ÐŸÐ°Ñ€Ð°Ð¼ÐµÑ‚Ñ€Ð¸
mu = 0
sigma = 1

# Ð“Ñ€Ð°Ñ„Ñ–ÐºÐ¸
t = np.linspace(-10, 10, 1000)
phi_mu0_sigma1 = phi(t, mu, sigma)
phi_mu1_sigma1 = phi(t, 1, sigma)
phi_mu0_sigma2 = phi(t, mu, 2*sigma)

plt.plot(t, phi_mu0_sigma1, label="$\mu = 0, \sigma = 1$")
plt.plot(t, phi_mu1_sigma1, label="$\mu = 1, \sigma = 1$")
plt.plot(t, phi_mu0_sigma2, label="$\mu = 0, \sigma = 2$")

plt.xlabel("$t$")
plt.ylabel("$\phi(t)$")
plt.legend()
mplcyberpunk.add_glow_effects()
plt.show()
```

```python
def phi(t, a):
  return np.exp(a * t)

# ÐŸÐ°Ñ€Ð°Ð¼ÐµÑ‚Ñ€Ð¸
a = 1

# Ð“Ñ€Ð°Ñ„Ñ–ÐºÐ¸
t = np.linspace(-10, 10, 1000)
phi_a1 = phi(t, a)

plt.plot(t, phi_a1, label="$a = 1$")

plt.xlabel("$t$")
plt.ylabel("$\phi(t)$")
plt.legend()
mplcyberpunk.add_glow_effects()
plt.show()
```
