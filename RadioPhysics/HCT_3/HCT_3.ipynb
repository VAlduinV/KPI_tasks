# Task 1

```python
import numpy as np
import matplotlib.pyplot as plt
import mplcyberpunk

plt.style.use("cyberpunk")

def B_xi(tau):
    # Кореляційна функція
    A = 1
    omega_o = 1
    return A**2 / 2 * np.cos(omega_o * tau)

def g_xi(omega):
    # Спектральна щільність
    A = 1
    omega_o = 1
    return A**2 / 2 * (3 + (omega - omega_o)**2 / 4 + (omega + omega_o)**2 / 4)

# Графік кореляційної функції
tau = np.arange(-10, 10, 0.01)
B_xi_values = B_xi(tau)
plt.plot(tau, B_xi_values, c="YELLOW")
plt.title("Кореляційна функція випадкового процесу $\\xi(t)$")
plt.xlabel("$\tau$")
plt.ylabel("$B_\\xi(\tau)$")
mplcyberpunk.add_glow_effects()
plt.show()

# Графік спектральної щільності
omega = np.arange(-5, 5, 0.01)
g_xi_values = g_xi(omega)
plt.plot(omega, g_xi_values, c="RED")
plt.title("Спектральна щільність випадкового процесу $\\xi(t)$")
plt.xlabel("$\omega$")
plt.ylabel("$g_\\xi(\omega)$")
mplcyberpunk.add_glow_effects()
plt.show()
```

# Task №3

```python
h_0 = 1

def B_ksi(tau):
    result = np.zeros_like(tau)
    non_zero_indices = np.abs(tau) <= h_0
    result[non_zero_indices] = 0.5
    return result

tau = np.arange(-10, 10, 0.1)
B_ksi_values = B_ksi(tau)

plt.plot(tau, B_ksi_values)
plt.xlabel("Затримка, $\\tau$")
plt.ylabel("Кореляційна функція, $B_{\\xi}(\\tau)$")
mplcyberpunk.add_glow_effects()
plt.show()

def g_ksi(w):
    result = np.zeros_like(w)
    non_zero_indices = np.abs(w) <= h_0
    result[non_zero_indices] = 1
    return result

w = np.arange(-10, 10, 0.1)
g_ksi_values = g_ksi(w)

plt.step(w, g_ksi_values, where='mid')  # Use step function for a clearer step-like plot
plt.xlabel("Частота, $w$")
plt.ylabel("Спектральна щільність, $g_{\\xi}(w)$")
mplcyberpunk.add_glow_effects()
plt.show()
```

# Task №4

```python
# Параметри процесу
a = 0.1

# Кореляційна функція
def B_xi(tau):
    return np.exp(-a * np.abs(tau))

# Спектральна щільність
def S_xi(w):
    return np.abs(B_xi(w))**2

# Графік кореляційної функції
tau = np.linspace(-10, 10, 1000)
B_xi_t = B_xi(tau)
plt.plot(tau, B_xi_t)
plt.xlabel("$t$")
plt.ylabel("$B_\\xi(t)$")
mplcyberpunk.add_glow_effects()
plt.show()

# Графік спектральної щільності
w = np.linspace(-5, 5, 1000)
S_xi_w = S_xi(w)
plt.plot(w, S_xi_w)
plt.xlabel("$w$")
plt.ylabel("$S_\\xi(w)$")
mplcyberpunk.add_glow_effects()
plt.show()
```

# Task №5

```python
# Задаємо параметри
delta = 1
w0 = 2

# Розраховуємо кореляційну функцію
def B(tau):
    return np.exp(-0.5 * tau**2 * delta**2) * np.cos(w0 * tau)

tau = np.arange(0, 10, 0.01)
B_tau = B(tau)

# Рисуємо графік
plt.plot(tau, B_tau)
plt.xlabel("Час, $t$")
plt.ylabel("Кореляційна функція, $B(t)$")
mplcyberpunk.add_glow_effects()
plt.show()

# Розраховуємо спектральну густину
def G(w):
    return delta * (1 - 1j * w0 * delta) / (w**2 + delta**2)

w = np.linspace(-5, 5, 1000)
G_w = G(w)

# Рисуємо графік
plt.plot(w, G_w)
plt.xlabel("Частота, $w$")
plt.ylabel("Спектральна густина, $G(w)$")
mplcyberpunk.add_glow_effects()
plt.show()
```

# Task №7

```python
import numpy as np
import matplotlib.pyplot as plt

# Кореляційні функції
def B_ksi_1(tau):
    return sigma**2 * np.exp(-a * np.abs(tau))

def B_ksi_2(tau):
    return sigma**2 * np.exp(-a * np.abs(tau)) * np.cos(w_o * tau)

# Спектральні щільності
def g_ksi_1(w):
    return sigma**2 * a / (a**2 + w**2)

def g_ksi_2(w):
    return sigma**2 * a / (a**2 + (w - w_o)**2) + sigma**2 * a / (a**2 + (w + w_o)**2)

# Параметри
sigma = 1
a = 1
w_o = 2 * np.pi

# Частотний діапазон
f_min = 0
f_max = 10
n = 1000
f = np.linspace(f_min, f_max, n)

# Обчислення спектрів
g_ksi_1_vals = g_ksi_1(f)
g_ksi_2_vals = g_ksi_2(f)

# Побудова графіків
plt.figure(figsize=(12, 8))

# Графік 1: Кореляційні функції
plt.subplot(2, 1, 1)
plt.plot(f, B_ksi_1(f), c="RED", label="$B_{\\xi_1}(\\tau)$")
plt.plot(f, B_ksi_2(f), c="BLUE", label="$B_{\\xi_2}(\\tau)$")
plt.xlabel("Затримка $\\tau$")
plt.ylabel("Кореляційна функція")
plt.legend()
mplcyberpunk.add_glow_effects()

# Графік 2: Спектральні щільності
plt.subplot(2, 1, 2)
plt.plot(f, g_ksi_1_vals, c="RED", label="Спектр $\\xi_1(t)$")
plt.plot(f, g_ksi_2_vals, c="BLUE", label="Спектр $\\xi_2(t)$")
plt.xlabel("Частота $\\omega$")
plt.ylabel("Спектральна щільність")
plt.legend()
mplcyberpunk.add_glow_effects()

plt.tight_layout()  # Додає автоматичний відступ між графіками
plt.show()
```

# Task №8

```python
# Ширина спектру
delta = 1

# Кореляційна функція
def correlation_function(tau):
    return 1 - (tau / (2 * delta))**2

# Спектр
def spectrum(w):
    return np.exp(-0.5 * w**2 / delta**2)

# Графік кореляційної функції
tau = np.arange(-10, 10, 0.01)
plt.plot(tau, correlation_function(tau), c="BLUE")
plt.xlabel("Таймлайн ($\\tau$)")
plt.ylabel("Кореляційна функція ($B(\\tau)$)")
plt.title("Кореляційна функція випадкового процесу")
mplcyberpunk.add_glow_effects()
plt.show()

# Графік спектру
w = np.arange(-10, 10, 0.01)
plt.plot(w, spectrum(w), c="RED")
plt.xlabel("Частота ($\\omega$)")
plt.ylabel("Спектральна густина ($G(\\omega)$)")
plt.title("Спектр випадкового процесу")
mplcyberpunk.add_glow_effects()
plt.show()
```
