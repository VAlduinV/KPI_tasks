# Task 1

```python
import numpy as np
import matplotlib.pyplot as plt
import mplcyberpunk

plt.style.use("cyberpunk")

def B_xi(tau):
    #  
    A = 1
    omega_o = 1
    return A**2 / 2 * np.cos(omega_o * tau)

def g_xi(omega):
    #  
    A = 1
    omega_o = 1
    return A**2 / 2 * (3 + (omega - omega_o)**2 / 4 + (omega + omega_o)**2 / 4)

#   
tau = np.arange(-10, 10, 0.01)
B_xi_values = B_xi(tau)
plt.plot(tau, B_xi_values, c="YELLOW")
plt.title("    $\\xi(t)$")
plt.xlabel("$\tau$")
plt.ylabel("$B_\\xi(\tau)$")
mplcyberpunk.add_glow_effects()
plt.show()

#   
omega = np.arange(-5, 5, 0.01)
g_xi_values = g_xi(omega)
plt.plot(omega, g_xi_values, c="RED")
plt.title("    $\\xi(t)$")
plt.xlabel("$\omega$")
plt.ylabel("$g_\\xi(\omega)$")
mplcyberpunk.add_glow_effects()
plt.show()
```

# Task 3

```python
h_0 = 1

def B_ksi(tau):
    result = np.zeros_like(tau)
    non_zero_indices = np.abs(tau) <= h_0
    result[non_zero_indices] = 0.5
    return result

tau = np.arange(-10, 10, 0.1)
B_ksi_values = B_ksi(tau)

plt.plot(tau, B_ksi_values)
plt.xlabel(", $\\tau$")
plt.ylabel(" , $B_{\\xi}(\\tau)$")
mplcyberpunk.add_glow_effects()
plt.show()

def g_ksi(w):
    result = np.zeros_like(w)
    non_zero_indices = np.abs(w) <= h_0
    result[non_zero_indices] = 1
    return result

w = np.arange(-10, 10, 0.1)
g_ksi_values = g_ksi(w)

plt.step(w, g_ksi_values, where='mid')  # Use step function for a clearer step-like plot
plt.xlabel(", $w$")
plt.ylabel(" , $g_{\\xi}(w)$")
mplcyberpunk.add_glow_effects()
plt.show()
```

# Task 5

```python
#  
delta = 1
w0 = 2

#   
def B(tau):
    return np.exp(-0.5 * tau**2 * delta**2) * np.cos(w0 * tau)

tau = np.arange(0, 10, 0.01)
B_tau = B(tau)

#  
plt.plot(tau, B_tau)
plt.xlabel(", $t$")
plt.ylabel(" , $B(t)$")
mplcyberpunk.add_glow_effects()
plt.show()

#   
def G(w):
    return delta * (1 - 1j * w0 * delta) / (w**2 + delta**2)

w = np.linspace(-5, 5, 1000)
G_w = G(w)

#  
plt.plot(w, G_w)
plt.xlabel(", $w$")
plt.ylabel(" , $G(w)$")
mplcyberpunk.add_glow_effects()
plt.show()
```

# Task 7

```python
import numpy as np
import matplotlib.pyplot as plt

#  
def B_ksi_1(tau):
    return sigma**2 * np.exp(-a * np.abs(tau))

def B_ksi_2(tau):
    return sigma**2 * np.exp(-a * np.abs(tau)) * np.cos(w_o * tau)

#  
def g_ksi_1(w):
    return sigma**2 * a / (a**2 + w**2)

def g_ksi_2(w):
    return sigma**2 * a / (a**2 + (w - w_o)**2) + sigma**2 * a / (a**2 + (w + w_o)**2)

# 
sigma = 1
a = 1
w_o = 2 * np.pi

#  
f_min = 0
f_max = 10
n = 1000
f = np.linspace(f_min, f_max, n)

#  
g_ksi_1_vals = g_ksi_1(f)
g_ksi_2_vals = g_ksi_2(f)

#  
plt.figure(figsize=(12, 8))

#  1:  
plt.subplot(2, 1, 1)
plt.plot(f, B_ksi_1(f), c="RED", label="$B_{\\xi_1}(\\tau)$")
plt.plot(f, B_ksi_2(f), c="BLUE", label="$B_{\\xi_2}(\\tau)$")
plt.xlabel(" $\\tau$")
plt.ylabel(" ")
plt.legend()
mplcyberpunk.add_glow_effects()

#  2:  
plt.subplot(2, 1, 2)
plt.plot(f, g_ksi_1_vals, c="RED", label=" $\\xi_1(t)$")
plt.plot(f, g_ksi_2_vals, c="BLUE", label=" $\\xi_2(t)$")
plt.xlabel(" $\\omega$")
plt.ylabel(" ")
plt.legend()
mplcyberpunk.add_glow_effects()

plt.tight_layout()  #     
plt.show()
```
