```python
import numpy as np
import matplotlib.pyplot as plt
import mplcyberpunk

plt.style.use("cyberpunk")

#  
data = [
    ["0-4", 84577, 78372, 79095, 73088],
    ["5-9", 87734, 82228, 89429, 83536],
    ["10-14", 78836, 75335, 81257, 77556],
    ["15-19", 72361, 69801, 75327, 72297],
    ["20-24", 74196, 68073, 69597, 63235],
    ["25-29", 90699, 90319, 87991, 86831],
    ["30-34", 112840, 124754, 107666, 118908],
    ["35-39", 136017, 148883, 134835, 148985],
    ["40-44", 116716, 132215, 119307, 135863],
    ["45-49", 102315, 114615, 105139, 118491],
    ["50-54", 79808, 91286, 83095, 94737],
    ["55-59", 77019, 102583, 74581, 97507],
    ["60-64", 73663, 109107, 73732, 108830],
    ["65-69", 62543, 101666, 61898, 101663],
    ["70-74", 46408, 78925, 48868, 84577],
    ["75-79", 23501, 43347, 22517, 42193],
    ["80", 30323, 62398, 30341, 62962]
]

#    
data = np.array(data)

#   
total_population_2020 = data[:, 1].astype(int)

#   
births = data[:, 3].astype(int) - data[:, 1].astype(int)

#   (,      )
population_projection_base = np.zeros((51, len(data)))
population_projection_base[0] = total_population_2020
for year in range(1, 51):
    population_projection_base[year] = population_projection_base[year - 1] + births

#   (     )
population_projection_optimistic = np.zeros((51, len(data)))
population_projection_optimistic[0] = total_population_2020
for year in range(1, 51):
    population_projection_optimistic[year] = population_projection_optimistic[year - 1] + births * 1.02  # 2%  

#   (     )
population_projection_pessimistic = np.zeros((51, len(data)))
population_projection_pessimistic[0] = total_population_2020
for year in range(1, 51):
    population_projection_pessimistic[year] = population_projection_pessimistic[year - 1] + births * 0.98  # 2%  

#       
age_groups = data[:, 0]
plt.figure(figsize=(12, 6))
plt.bar(age_groups, births)
plt.xlabel(" ")
plt.ylabel(" ")
plt.title("     (2020-2021). Ivanov. V.V.")
plt.xticks(rotation=45)
plt.show()

#        
years = range(2020, 2071)
plt.figure(figsize=(12, 6))
for i in range(len(data)):
    plt.plot(years, population_projection_base[:, i], label=age_groups[i])
plt.xlabel("")
plt.ylabel(" ")
plt.title("      ( ). Ivanov V.V.")
plt.legend()
plt.grid(True)
mplcyberpunk.add_gradient_fill(alpha_gradientglow=0.15)
plt.show()

#        
plt.figure(figsize=(12, 6))
for i in range(len(data)):
    plt.plot(years, population_projection_optimistic[:, i], label=age_groups[i])
plt.xlabel("")
plt.ylabel(" ")
plt.title("      ( ). Ivanov V.V.")
plt.legend()
plt.grid(True)
mplcyberpunk.add_gradient_fill(alpha_gradientglow=0.15)
plt.show()

#        
plt.figure(figsize=(12, 6))
for i in range(len(data)):
    plt.plot(years, population_projection_pessimistic[:, i], label=age_groups[i])
plt.xlabel("")
plt.ylabel(" ")
plt.title("      ( ). Ivanov V.V.")
plt.legend()
plt.grid(True)
mplcyberpunk.add_gradient_fill(alpha_gradientglow=0.15)
plt.show()
```
